```python
import tkinter as tk

def on_button_click(event):
    text = event.widget.cget("text")
    
    if text == "=":
        try:
            result = str(eval(entry.get()))
            entry.delete(0, tk.END)
            entry.insert(tk.END, result)
        except Exception as e:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif text == "C":
        entry.delete(0, tk.END)
    else:
        entry.insert(tk.END, text)

root = tk.Tk()
root.geometry("300x400")
root.title("Calculator")

entry = tk.Entry(root, font=("Helvetica", 20))
entry.pack(fill=tk.BOTH, ipadx=8, padx=10, pady=10)

button_frame = tk.Frame(root)
button_frame.pack()

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    'C', '0', '=', '+',
    
]

row_value = 1
col_value = 0

for button in buttons:
    tk_button = tk.Button(button_frame, text=button, font=("Helvetica", 15), padx=20, pady=20)
    tk_button.grid(row=row_value, column=col_value)
    tk_button.bind("<Button-1>", on_button_click)
    
    col_value += 1
    if col_value > 3:
        col_value = 0
        row_value += 1

root.mainloop()
```

```python
import tkinter as tk
import string
import random

def generate_password():
    password_length = int(length_entry.get())
    
    if password_length <= 0:
        password_result.config(text="Invalid length")
        return
    
    characters = string.ascii_letters + string.digits + string.punctuation
    generated_password = ''.join(random.choice(characters) for _ in range(password_length))
    
    password_result.config(text=generated_password)

def accept_password():
    name = name_entry.get()
    password = password_result.cget("text")
    
    if name and password:
        with open("passwords.txt", "a") as file:
            file.write(f"Name: {name}\tPassword: {password}\n")
        
        name_entry.delete(0, tk.END)
        password_result.config(text="")
        status_label.config(text="Password saved successfully!", fg="green")
    else:
        status_label.config(text="Please generate a password first.", fg="red")

def reset_password():
    name_entry.delete(0, tk.END)
    length_entry.delete(0, tk.END)
    password_result.config(text="")
    status_label.config(text="")

root = tk.Tk()
root.geometry("400x350")
root.title("Password Generator")
root.configure(bg="white")

title_label = tk.Label(root, text="Password Generator", font=("Helvetica", 16, "bold"), fg="blue", bg="white")
title_label.pack(pady=10)

name_label = tk.Label(root, text="Enter User Name:", font=("Helvetica", 12), bg="white")
name_label.pack()

name_entry = tk.Entry(root, font=("Helvetica", 12))
name_entry.pack()

length_label = tk.Label(root, text="Enter Password Length:", font=("Helvetica", 12), bg="white")
length_label.pack()

length_entry = tk.Entry(root, font=("Helvetica", 12))
length_entry.pack()

generate_button = tk.Button(root, text="Generate Password", font=("Helvetica", 12), bg="blue", fg="white", command=generate_password)
generate_button.pack(pady=10)

password_result = tk.Label(root, text="", font=("Helvetica", 12), bg="white")
password_result.pack()

accept_button = tk.Button(root, text="Accept", font=("Helvetica", 12), bg="white", fg="blue", command=accept_password)
accept_button.pack(pady=5)

status_label = tk.Label(root, text="", font=("Helvetica", 12), bg="white")
status_label.pack()

reset_button = tk.Button(root, text="Reset", font=("Helvetica", 12), bg="white", fg="blue", command=reset_password)
reset_button.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
import tkinter.simpledialog

class TodoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List")
        self.root.geometry("500x400")
        
        self.tasks = []
        
        self.title_label = tk.Label(root, text="To-Do List", font=("Times", 24, "bold italic"), fg="black", bg="green")
        self.title_label.pack(fill=tk.X, padx=10, pady=10)
        
        self.add_title_label = tk.Label(root, text="Add Items", font=("Helvetica", 12))
        self.add_title_label.pack(padx=10, anchor="w")
        
        self.add_frame = tk.Frame(root)
        self.add_frame.pack(padx=10, fill=tk.BOTH)
        
        self.add_entry = tk.Entry(self.add_frame, font=("Helvetica", 12))
        self.add_entry.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        
        self.add_button = tk.Button(self.add_frame, text="Add", font=("Helvetica", 12), bg="blue", fg="white", command=self.add_task)
        self.add_button.pack(side=tk.LEFT)
        
        self.tasks_title_label = tk.Label(root, text="Tasks", font=("Helvetica", 12))
        self.tasks_title_label.pack(padx=10, pady=(20, 0), anchor="w")
        
        self.tasks_frame = tk.Frame(root)
        self.tasks_frame.pack(fill=tk.BOTH, padx=10, pady=10, expand=True)
        
        self.display_tasks()
    
    def add_task(self):
        task_text = self.add_entry.get()
        if task_text:
            self.tasks.append(task_text)
            self.add_entry.delete(0, tk.END)
            self.display_tasks()
    
    def edit_task(self, index):
        new_text = tk.simpledialog.askstring("Edit Task", "Edit the task:", initialvalue=self.tasks[index])
        if new_text:
            self.tasks[index] = new_text
            self.display_tasks()
    
    def delete_task(self, index):
        del self.tasks[index]
        self.display_tasks()
    
    def display_tasks(self):
        for widget in self.tasks_frame.winfo_children():
            widget.destroy()
        
        for index, task in enumerate(self.tasks):
            task_frame = tk.Frame(self.tasks_frame, bd=2, relief="groove")
            task_frame.grid(row=index, column=0, pady=(5, 0), sticky="we")
            
            task_label = tk.Label(task_frame, text=task, font=("Helvetica", 12), anchor="w")
            task_label.pack(side=tk.LEFT, padx=5)
            
            edit_button = tk.Button(task_frame, text="Edit", font=("Helvetica", 12),bg='green',fg='white', command=lambda i=index: self.edit_task(i))
            edit_button.pack(side=tk.LEFT, padx=10)
            
            delete_button = tk.Button(task_frame, text="Delete", font=("Helvetica", 12),bg='red',fg='white', command=lambda i=index: self.delete_task(i))
            delete_button.pack(side=tk.LEFT, padx=10)
    
if __name__ == "__main__":
    root = tk.Tk()
    app = TodoApp(root)
    root.mainloop()
```
